```python
df['Author'].value_counts()
```

```python
a = df.groupby(['Author', 'Genre']).agg({'Name': 'count'}).unstack()['Name', 'Non Fiction'].sort_values(ascending=False)[:11]
b = df.groupby(['Author', 'Genre']).agg({'Name': 'count'}).unstack()['Name', 'Fiction'].sort_values(ascending=False)[:11]
```

```python
a
```

```python
b
```

```python
with plt.style.context('dark_background'):
    figure, axis = plt.subplots(1, 2, figsize=(5, 5))
    
    axis[0].set_xlabel('Number of appreances')
    axis[0].set_title('Authors-Non Fiction-')
    axis[0].barh(y=a.index, width=a.values)
    axis[0].invert_xaxis()
    axis[0].yaxis.tick_left()
    axis[0].set_xticks(np.arange(max(a.values)+1))
    axis[0].set_yticklabels(a.index)
    
    axis[1].set_title('Authors-Fiction-')
    axis[1].set_xlabel('Number of appreances')
    axis[1].barh(y=b.index, width=b.values)
    axis[1].yaxis.tick_right()
    axis[1].set_xticks(np.arange(max(b.values)+1))
    axis[1].set_yticklabels(b.index)
    
plt.show()
```

```python
price = pd.cut(df['Price'], bins=15)
price = pd.DataFrame(pd.DataFrame(price).groupby('Price').size(), columns=['Price Count']).reset_index()
user_rating = pd.cut(df['User Rating'], bins=15)
user_rating = pd.DataFrame(pd.DataFrame(user_rating).groupby('User Rating').size(), columns=['User Rating Count']).reset_index()
reviews = pd.cut(df['Reviews'], bins=15)
reviews = pd.DataFrame(pd.DataFrame(reviews).groupby('Reviews').size(), columns=['Reviews Count']).reset_index()

display(pd.concat([price, user_rating, reviews], axis=1))
```
